## Carga y exploración de datos

### Subtask:
Cargar el conjunto de datos "Pima Indians Diabetes Dataset" desde la URL proporcionada, explorar su estructura, variables y describir brevemente su relación con la predicción de diabetes.


**Reasoning**:
Se importará la librería pandas y se cargará el conjunto de datos desde la URL proporcionada en un DataFrame, asignando nombres a las columnas. Luego, se mostrarán las primeras filas, información general y un resumen estadístico del DataFrame para explorar los datos. Finalmente, se contará el número de valores en la columna 'resultado'.



```python
import pandas as pd

# Nombres de las columnas
nombres_columnas = ['embarazos', 'glucosa', 'presion_arterial', 'espesor_piel', 'insulina', 'bmi', 'pedigree', 'edad', 'resultado']

# Cargar el conjunto de datos
df = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv", names=nombres_columnas)

# Mostrar las primeras 5 filas
print("Primeras 5 filas del DataFrame:")
display(df.head())

# Mostrar información del DataFrame
print("\nInformación del DataFrame:")
display(df.info())

# Mostrar resumen estadístico
print("\nResumen estadístico del DataFrame:")
display(df.describe())

# Contar valores en la columna 'resultado'
print("\nConteo de valores en la columna 'resultado':")
display(df['resultado'].value_counts())
```

## Preprocesamiento de datos

### Subtask:
Escalar las variables numéricas utilizando `StandardScaler` y dividir el conjunto de datos en conjuntos de entrenamiento y prueba.


**Reasoning**:
Separar las características y la variable objetivo, escalar las características numéricas y dividir los datos en conjuntos de entrenamiento y prueba.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Separar características (X) y variable objetivo (y)
X = df.drop('resultado', axis=1)
y = df['resultado']

# Instanciar StandardScaler
scaler = StandardScaler()

# Aplicar StandardScaler a las características
X_scaled = scaler.fit_transform(X)

# Dividir los datos escalados en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

print("Dimensiones de los conjuntos de datos:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

## Modelo base

### Subtask:
Entrenar un `RandomForestClassifier` con los hiperparámetros por defecto y evaluar su rendimiento utilizando métricas como precisión, recall, F1-Score y AUC.


* Importar las bibliotecas necesarias para el entrenamiento y evaluación del modelo.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score
```

* Crear una instancia y entrenamiento de un RandomForestClassifier con hiperparámetros predeterminados y realización de predicciones en el conjunto de prueba.



```python
# Instanciar un RandomForestClassifier con hiperparámetros por defecto
rf_model = RandomForestClassifier(random_state=42)

# Entrenar el modelo utilizando los datos de entrenamiento escalados
rf_model.fit(X_train, y_train)

# Realizar predicciones sobre el conjunto de prueba escalado
y_pred = rf_model.predict(X_test)
```

**Reasoning**:
Calculate and print the evaluation metrics for the trained Random Forest model using the test set predictions and the true test labels.



```python
# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Imprimir las métricas
print(f"Precisión del modelo: {accuracy:.4f}")
print(f"F1-Score del modelo: {f1:.4f}")
print(f"Recall del modelo: {recall:.4f}")
print(f"AUC del modelo: {auc:.4f}")
```

## Aplicar grid search

### Subtask:
Definir un rango de hiperparámetros para `RandomForestClassifier` y utilizar `GridSearchCV` para encontrar la mejor combinación de hiperparámetros basándose en el rendimiento del modelo.


**Reasoning**:
Import the GridSearchCV class to perform hyperparameter tuning.



```python
from sklearn.model_selection import GridSearchCV
```

**Reasoning**:
Define the hyperparameter grid, instantiate the RandomForestClassifier, and instantiate and fit GridSearchCV to the training data.



```python
# Definir el rango de hiperparámetros a tunear
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Instanciar un RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Instanciar GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1)

# Ajustar GridSearchCV a los datos de entrenamiento escalados
grid_search.fit(X_train, y_train)
```

**Reasoning**:
Print the best hyperparameters and the best F1-score found by GridSearchCV, and store the best estimator.



```python
# Imprimir los mejores hiperparámetros
print("Mejores hiperparámetros encontrados por GridSearchCV:")
print(grid_search.best_params_)

# Imprimir el mejor score (F1-score) obtenido por GridSearchCV
print("\nMejor F1-score obtenido por GridSearchCV:")
print(grid_search.best_score_)

# Almacenar el mejor modelo
best_grid_model = grid_search.best_estimator_
```

## Aplicar random search

### Subtask:
Definir un rango más amplio de hiperparámetros y utilizar `RandomizedSearchCV` con un número específico de iteraciones para encontrar una buena combinación de hiperparámetros.


**Reasoning**:
Import the necessary class for randomized search and define a broader hyperparameter distribution for the RandomForestClassifier.



```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Definir un rango más amplio de hiperparámetros
param_dist = {
    'n_estimators': randint(low=50, high=500),
    'max_depth': [None] + list(randint(low=5, high=50).rvs(10)), # None and 10 random integers between 5 and 50
    'min_samples_split': randint(low=2, high=20)
}
```

**Reasoning**:
Instantiate a RandomForestClassifier and RandomizedSearchCV, then fit the search object to the training data.



```python
# Instanciar un RandomForestClassifier con un random_state para reproducibilidad
rf = RandomForestClassifier(random_state=42)

# Instanciar RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=100, cv=5, scoring='f1', n_jobs=-1, random_state=42)

# Ajustar RandomizedSearchCV a los datos de entrenamiento escalados
random_search.fit(X_train, y_train)
```

**Reasoning**:
Print the best hyperparameters and best score found by RandomizedSearchCV and store the best model.



```python
# Imprimir los mejores hiperparámetros encontrados por RandomizedSearchCV
print("Mejores hiperparámetros encontrados por RandomizedSearchCV:")
print(random_search.best_params_)

# Imprimir el mejor score (F1-score) obtenido por RandomizedSearchCV
print("\nMejor F1-score obtenido por RandomizedSearchCV:")
print(random_search.best_score_)

# Almacenar el mejor modelo encontrado por RandomizedSearchCV
best_random_model = random_search.best_estimator_
```

## Comparación y análisis

### Subtask:
Evaluar los modelos obtenidos con Grid Search y Random Search en el conjunto de prueba, comparar sus métricas de rendimiento y los tiempos de entrenamiento.


**Reasoning**:
Import the necessary libraries for evaluation and time measurement.



```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score
import time
```

**Reasoning**:
Define a function to train and evaluate a model, then call it for each model.



```python
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    """Entrena y evalúa un modelo de clasificación, imprimiendo las métricas."""
    print(f"Evaluando modelo: {model_name}")

    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    training_time = end_time - start_time
    print(f"Tiempo de entrenamiento: {training_time:.4f} segundos")

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)

    print(f"Precisión: {accuracy:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"AUC: {auc:.4f}")
    print("-" * 30)

    return {'accuracy': accuracy, 'f1': f1, 'recall': recall, 'auc': auc, 'training_time': training_time}

# Evaluar el modelo base
base_model_results = train_and_evaluate_model(rf_model, X_train, y_train, X_test, y_test, "Modelo Base (RandomForest por defecto)")

# Evaluar el mejor modelo de Grid Search
grid_search_model_results = train_and_evaluate_model(best_grid_model, X_train, y_train, X_test, y_test, "Mejor Modelo (Grid Search)")

# Evaluar el mejor modelo de Random Search
random_search_model_results = train_and_evaluate_model(best_random_model, X_train, y_train, X_test, y_test, "Mejor Modelo (Random Search)")
```

## Reflexión final

### Subtask:
Analizar cuál técnica de búsqueda de hiperparámetros fue más eficiente, cuál encontró el mejor modelo y reflexionar sobre posibles enfoques alternativos.


**Reasoning**:
Analizaré los resultados comparando las métricas de rendimiento y los tiempos de entrenamiento de los tres modelos para determinar la eficiencia de las técnicas de búsqueda de hiperparámetros y discutiré enfoques alternativos, como se indica en las instrucciones.



```python
import numpy as np

print("--- Análisis de los Resultados ---")

# 1. Comparar los resultados de los modelos
print("\nComparación de Métricas de Rendimiento y Tiempo de Entrenamiento:")
print("-----------------------------------------------------------------")
print(f"{'Modelo':<30} | {'Accuracy':<10} | {'F1-Score':<10} | {'Recall':<10} | {'AUC':<10} | {'Tiempo (s)':<15}")
print("-" * 100)
print(f"{'Modelo Base (RandomForest)':<30} | {base_model_results['accuracy']:<10.4f} | {base_model_results['f1']:<10.4f} | {base_model_results['recall']:<10.4f} | {base_model_results['auc']:<10.4f} | {base_model_results['training_time']:<15.4f}")
print(f"{'Mejor Modelo (Grid Search)':<30} | {grid_search_model_results['accuracy']:<10.4f} | {grid_search_model_results['f1']:<10.4f} | {grid_search_model_results['recall']:<10.4f} | {grid_search_model_results['auc']:<10.4f} | {grid_search_model_results['training_time']:<15.4f}")
print(f"{'Mejor Modelo (Random Search)':<30} | {random_search_model_results['accuracy']:<10.4f} | {random_search_model_results['f1']:<10.4f} | {random_search_model_results['recall']:<10.4f} | {random_search_model_results['auc']:<10.4f} | {random_search_model_results['training_time']:<15.4f}")

# 2. Analizar qué técnica encontró el mejor modelo
print("\nAnálisis del Mejor Modelo:")
print("---------------------------")

best_f1_grid = grid_search_model_results['f1']
best_f1_random = random_search_model_results['f1']

if best_f1_random > best_f1_grid:
    print("Random Search encontró un modelo con un F1-Score ligeramente mejor en el conjunto de prueba.")
elif best_f1_grid > best_f1_random:
     print("Grid Search encontró un modelo con un F1-Score ligeramente mejor en el conjunto de prueba.")
else:
    print("Grid Search y Random Search encontraron modelos con el mismo F1-Score en el conjunto de prueba.")

# También considerar otras métricas
if random_search_model_results['auc'] > grid_search_model_results['auc']:
    print("Random Search también obtuvo un AUC ligeramente mayor.")
elif grid_search_model_results['auc'] > random_search_model_results['auc']:
    print("Grid Search también obtuvo un AUC ligeramente mayor.")


# 3. Discutir la eficiencia computacional
print("\nEficiencia Computacional:")
print("-------------------------")
print(f"Tiempo de entrenamiento del modelo base: {base_model_results['training_time']:.4f} segundos")
print(f"Tiempo de entrenamiento del mejor modelo de Grid Search: {grid_search_model_results['training_time']:.4f} segundos")
print(f"Tiempo de entrenamiento del mejor modelo de Random Search: {random_search_model_results['training_time']:.4f} segundos")

if grid_search_model_results['training_time'] < random_search_model_results['training_time']:
    print("Grid Search fue considerablemente más rápido que Random Search en este caso particular.")
elif random_search_model_results['training_time'] < grid_search_model_results['training_time']:
     print("Random Search fue más rápido que Grid Search en este caso particular.")
else:
    print("Los tiempos de entrenamiento fueron similares.")

# 4. Reflexión sobre por qué una técnica superó a la otra (o no)
print("\nReflexión:")
print("-----------")
print("En este caso, Random Search logró un F1-Score y AUC marginalmente superiores en el conjunto de prueba, lo que sugiere que exploró combinaciones de hiperparámetros que llevaron a un modelo ligeramente mejor en este espacio de búsqueda y con este número de iteraciones.")
print("A pesar de encontrar un modelo marginalmente mejor, Random Search tuvo un tiempo de entrenamiento significativamente mayor que Grid Search.")
print("La superioridad de una técnica sobre otra puede depender del tamaño y la naturaleza del espacio de búsqueda, el número de iteraciones (en el caso de Random Search) y la distribución de los hiperparámetros óptimos dentro de ese espacio.")
print("Grid Search explora exhaustivamente todas las combinaciones en una cuadrícula definida, mientras que Random Search muestrea aleatoriamente, lo que puede ser más eficiente en espacios de alta dimensión o cuando los hiperparámetros óptimos no están alineados con una cuadrícula predefinida.")

# 5. Sugerir posibles enfoques alternativos
print("\nEnfoques Alternativos para Optimización de Hiperparámetros:")
print("----------------------------------------------------------")
print("- **Optimización Bayesiana:** Utiliza resultados de evaluaciones previas para guiar la búsqueda hacia regiones prometedoras del espacio de hiperparámetros.")
print("- **Optimización basada en Gradientes:** (Aunque menos común para hiperparámetros de modelos de árboles) Explora el espacio de hiperparámetros utilizando información de gradientes.")
print("- **Algoritmos Genéticos:** Emplean principios de selección natural para encontrar combinaciones óptimas de hiperparámetros.")
print("- **Bibliotecas de Optimización Automática (AutoML):** Herramientas más completas que automatizan el proceso de selección de modelo e hiperparámetros (ej. Auto-Sklearn, TPOT).")

# 6. Resumen conciso
print("\n--- Resumen ---")
print("En este análisis, comparamos los modelos de RandomForest obtenidos con hiperparámetros por defecto, Grid Search y Random Search.")
print("Grid Search fue significativamente más rápido en el entrenamiento, mientras que Random Search encontró un modelo con métricas de rendimiento (F1-Score, AUC) marginalmente mejores en el conjunto de prueba.")
print("Esto sugiere que, aunque Random Search puede encontrar modelos ligeramente mejores al explorar un espacio de búsqueda más amplio, puede ser menos eficiente computacionalmente dependiendo del número de iteraciones y el espacio definido.")
print("Existen enfoques alternativos como la Optimización Bayesiana y las herramientas de AutoML que podrían explorarse para una optimización más eficiente y potencialmente mejor.")
```
